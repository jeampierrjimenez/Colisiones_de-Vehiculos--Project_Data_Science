```python
df_vehicle_crash_clean.columns
```

NUMBER_OF_PERSONS_KILLED, NUMBER_OF_PEDESTRIANS_KILLED, NUMBER_OF_CYCLIST_KILLED, NUMBER_OF_MOTORIST_KILLED

```python
df_death_perc = df_vehicle_crash_clean[["COLLISION_ID", "NUMBER_OF_PERSONS_KILLED", "NUMBER_OF_PEDESTRIANS_KILLED", "NUMBER_OF_CYCLIST_KILLED", "NUMBER_OF_MOTORIST_KILLED"]]
```

```python
df_death_perc = df_death_perc[df_death_perc["NUMBER_OF_PERSONS_KILLED"] > 0]

df_death_perc #en este dataframe tengo todos los registros en los que murio ALGUIEN

dicc_death = dict()

dicc_death["TOT_CRASHES"] = registros
dicc_death["TOT_DEATHS"] = len(df_death_perc["NUMBER_OF_PERSONS_KILLED"])

dicc_death
```

vemos que el porcentaje del total de los choques que resultaron en al menos una muerte es de un 0.13%, siendo 2537 choques que resultaron en muerte de al menos un participante

```python
dicc_death["PERC_DEATH"] = ( (dicc_death["TOT_DEATHS"] * 100) / dicc_death["TOT_CRASHES"])

dicc_death
```

```python
df_death_perc = pd.DataFrame.from_dict(dicc_death, orient = "index").rename(columns = {0 : "VALUE"})
df_death_perc.round(3)
```

```python
df_climate_perc = df_vehicle_crash_clean[ (df_vehicle_crash_clean["CONTRIBUTING_FACTOR_VEHICLE_1"] == "Pavement Slippery") | (df_vehicle_crash_clean["CONTRIBUTING_FACTOR_VEHICLE_2"] == "Pavement Slippery") | (df_vehicle_crash_clean["CONTRIBUTING_FACTOR_VEHICLE_3"] == "Pavement Slippery") | (df_vehicle_crash_clean["CONTRIBUTING_FACTOR_VEHICLE_4"] == "Pavement Slippery") | (df_vehicle_crash_clean["CONTRIBUTING_FACTOR_VEHICLE_5"] == "Pavement Slippery")]
```
